```python
import pandas as pd

data = {
    'Candidate': [
        'Alice',
        'Bob',
        'Charlie',
        'David',
        'Emma'
    ],

    'Skills': [
        'Python SQL Machine Learning',
        'Python Excel Communication',
        'Java SQL Data Analysis',
        'Python Machine Learning Deep Learning',
        'SQL Excel Power BI'
    ]
}

df = pd.DataFrame(data)

print(df)
```

```python
job_skills = ['Python', 'Machine Learning', 'SQL']

def calculate_match(skills):
    candidate_skills = skills.split()
    matched = len(set(candidate_skills) & set(job_skills))
    return matched

df['Match Score'] = df['Skills'].apply(calculate_match)

print(df[['Candidate', 'Match Score']])
```

```python
ranked_df = df.sort_values(by='Match Score', ascending=False)

print("Candidate Ranking:")
print(ranked_df[['Candidate', 'Match Score']])
```

```python
job_skills = ['Python', 'Machine Learning', 'SQL']

def find_missing_skills(skills):
    candidate_skills = skills.split()
    missing = set(job_skills) - set(candidate_skills)
    return ', '.join(missing)

df['Missing Skills'] = df['Skills'].apply(find_missing_skills)

print(df[['Candidate', 'Missing Skills']])
```
